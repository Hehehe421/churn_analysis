```python
#Replace spaces with null in TotalCharges
data['TotalCharges'] =  data['TotalCharges'].replace(" ", np.nan)

#Check for missing values
n_missing = data[data['TotalCharges'].isnull()].shape[0]
percent_missing = round(data[data['TotalCharges'].isnull()].shape[0]/data.shape[0]*100,2)
print("There are", n_missing, 'records contains missing values, which is', percent_missing, '% of whole dataset')
```

```python
#Removing the 11 missing records 
data = data[data['TotalCharges'].notnull()]
```

```python
print ("Rows     : " ,data.shape[0])
print ("Columns  : " ,data.shape[1])
```

```python
#Convert the TotalCharges from character into numeric
data["TotalCharges"] = data["TotalCharges"].astype(float)
#data['Total'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
#Create numeric bucket
#1. tenure_group
def tenure_group_fun(x):
    if x['tenure'] <= 12:
        return 'tenure_group1'  
    if x['tenure'] <= 24:
        return 'tenure_group2'
    if x['tenure'] <= 36:
        return 'tenure_group2' 
    if x['tenure'] <= 48:
        return 'tenure_group3'
    if x['tenure'] <= 60:
        return 'tenure_group4'  
    if x['tenure'] <= 72:
        return 'tenure_group5'
    
data['tenure_group'] = data.apply(tenure_group_fun, axis = 1)
#2. MonthlyCharge cut by quantile into 4 group
bin_labels = ['monthly_Q1', 'monthly_Q2', 'monthly_Q3', 'monthly_Q4']
data['monthly_group'] = pd.qcut(data['MonthlyCharges'], q=[0, .25, .5, .75, 1], labels=bin_labels)

#3. TotalCharge cut by quantile into 4 group
bin_labels = ['total_Q1', 'total_Q2', 'total_Q3', 'total_Q4']
data['total_group'] = pd.qcut(data['TotalCharges'], q=[0, .25, .5, .75, 1], labels=bin_labels)
```

```python
def view_group(input_dataframe, view_feature):
    c = input_dataframe[view_feature].value_counts()
    p = round(input_dataframe[view_feature].value_counts(normalize = True)*100, 2)
    combine = pd.concat([c,p], axis = 1, keys = ['Count', '%']).sort_index()
    print('The bucketed new group summary: ')
    print(combine)
    print()
    print('The bucketd new group bar chart:')
    figure = plt.figure()
    input_dataframe[view_feature].value_counts().sort_index().plot(kind = 'bar')
    plt.xticks(rotation=15)
    plt.show()
    
view_group(data, 'tenure_group')
```

```python
#Separating catagorical and numerical columns
Id_col     = ['customerID']
target_col = ['Churn']
cat_cols   = data.nunique()[data.nunique() < 6].keys().tolist()
cat_cols   = [x for x in cat_cols if x not in target_col]
num_cols   = [x for x in data.columns if x not in cat_cols + target_col + Id_col]
print('The ID column is ', Id_col)
print('-----------------------------------------------------------------------------------------------------------------------')
print('The Target column is ', target_col)
print('-----------------------------------------------------------------------------------------------------------------------')
print('The Numerical columns are:')
print(cat_cols)
print('-----------------------------------------------------------------------------------------------------------------------')
print('The Catagorical columns are:')
print(num_cols)
```

```python
#Data Manipulation
#Convert Target Yes/No into 1/0
data['Churn'] = np.where(data.Churn =='Yes', 1, 0)
#One-hot encode and combine, drop the original feature and clean the value

def encode_and_bind(input_dataframe, feature_to_encode):
    #Create the dummies for given feature
    dummies = pd.get_dummies(input_dataframe[[feature_to_encode]])
    #Concat the dummies features
    res = pd.concat([input_dataframe, dummies], axis=1)
    #Drop the original one
    res = res.drop([feature_to_encode], axis=1)
    return(res) 
```

```python
for feature in cat_cols:
    data = encode_and_bind(data, feature)
```
